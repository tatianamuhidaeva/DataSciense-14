Текст отзывов пока использовать не будем. Выделим в отдельный столбец список дат последних двух отзывов Dates_Reviews.

На случай заполнения пропусков в признаке Количество отзывов, создадим признак: минимальное количество пропусков Min_count_Reviews

```python
pattern = re.compile('\d\d\/\d\d\/\d\d\d\d')

s_dates_rews = data['Reviews'].apply(lambda x: (pattern.findall(str(x))))
s_dates_rews = s_dates_rews.apply(lambda l: [datetime.strptime(d,'%m/%d/%Y') for d in l])
s_min_count_rews = s_dates_rews.apply(lambda l: len(l))
```

Получим дату публикации самого последнего отзыва. Ее будем считать датой выгрузки данных. latest_date  

```python
latest_date = datetime(2000,1,1,0,0,0)
for l_d in s_dates_rews:
    if len(l_d)>0:
        if latest_date<max(l_d):
            latest_date = max(l_d)
            
earest_date = datetime.now()
for l_d in s_dates_rews:
    if len(l_d)>0:
        if earest_date>min(l_d):
            earest_date = min(l_d)            
print(latest_date, earest_date)
```

Теперь получим два признака: Разница в днях между latest_date и последним отзывом ресторана, и разницу в днях между последними  двумя отзывами.

Если отзывов нет, или только один отзыв, то за дату первого отзыва берем latest_date, а за дату второго отзыва берем earest_date

На базе этих двух признаков введем признак Review_Frequency - частота отзывов. Естественно, что признак имеет большую погрешность, так как основан на малом количестве данных. Но это даст нам информацию для заполнения пропусков в Number of Reviews

```python
def get_delta_with_last(row):
    days = latest_date
    if (row['count'] > 0):
        days = latest_date - row['dates'][0]
    else:
        if (row['number'] == 0):
            days = latest_date - earest_date
        else:
            days = (latest_date - earest_date)/(row['number']+1)
    return days.days

def get_delta_between_two(row):
    days = latest_date
    if (row['count'] >= 2):
        days = row['dates'][0] - row['dates'][1]
    elif (row['count'] == 1):
        if (row['number'] == row['count']):
            days = row['dates'][0] - earest_date
        else:
            days = (row['dates'][0] - earest_date)/(row['number'])
    else:
        if (row['number'] == 0):
            days = latest_date - earest_date
        else:
            days = (latest_date - earest_date)/(row['number']+1)
    return days.days

d = pd.DataFrame({'number': data['Number of Reviews'],'count': s_min_count_rews, 'dates':s_dates_rews}) 
# Заполним пропуски Number of Reviews сначала 0, потом заменим 0 числом из Count
d['number'].fillna(0, inplace=True)
d['number'] = d.apply(lambda row: row['count'] if row['number']<row['count'] else row['number'], axis=1)
d['delta_with_last'] = d.apply(get_delta_with_last, axis=1);
d['delta_between_two'] = d.apply(get_delta_between_two, axis=1)
d['frequence'] = d.apply(lambda row: (row['delta_with_last']+row['delta_between_two'])/2, axis = 1)
```

```python
data['Reviews_frequence'] = d['frequence']
data['Number of Reviews'] = d['number']
data['delta_with_last'] = d['delta_with_last']
data['delta_between_two'] = d['delta_between_two']
```

# Data Preprocessing
Теперь, для удобства и воспроизводимости кода, завернем всю обработку в одну большую функцию.

```python
# на всякий случай, заново подгружаем данные
df_train = pd.read_csv(DATA_DIR+'/main_task.csv')
df_test = pd.read_csv(DATA_DIR+'/kaggle_task.csv')
df_worldcities = pd.read_csv('/kaggle/input/world-cities-database/worldcitiespop.csv')
df_train['sample'] = 1 # помечаем где у нас трейн
df_test['sample'] = 0 # помечаем где у нас тест
df_test['Rating'] = 0 # в тесте у нас нет значения Rating, мы его должны предсказать, по этому пока просто заполняем нулями

data = df_test.append(df_train, sort=False).reset_index(drop=True) # объединяем
data.info()
```

```python
import math
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import re # for search on pattern
from datetime import datetime
from datetime import timedelta

import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline

# Загружаем специальный удобный инструмент для разделения датасета:
from sklearn.model_selection import train_test_split

class RestaurantsDF():
        def __init__(self, data):
            self.data = data
        
        def show_head50(self):
            display(self.data.head(50))
        
        def show_tail50(self):
            display(self.data.tail(50))
        
        def show_sample50(self):
            display(self.data.sample(50))
            
        def show_info(self):
            display(self.data.info())
            
        def proc_Price_Range(self):
            # Обработка Price_Range            
            # Обработка пустых значений и замена строк на числа
            data = self.data.copy()
            data['Price_Range_isNAN'] = pd.isna(data['Price Range']).astype('uint8')
            if ('$' in data['Price Range'].unique()):
                data['Price Range'] = data['Price Range'].apply(lambda x: 1 if (str(x) == '$') else 2 if (str(x) == '$$ - $$$') else 3 if (str(x) == '$$$$') else 2)
            self.data = data

        def clear_name(val):
            if pd.isna(val): return 'unknown'
            val = val.strip('[]')
            val = val.replace("'",'')
            val = val.split(', ')
            return val
        
        def proc_Cuisine(self):
            data = self.data.copy()
            # Пустые значения заполним словом 'nothing'
            data['Cuisine Style'] = data['Cuisine Style'].fillna("['nothing']")
            data['Cuisine Style'] = data['Cuisine Style'].apply(clear_name)
            
            # Создаем новый признак - Кол-во Кухонь в ресторане
            data['count_Cuisine'] = data['Cuisine Style'].apply(lambda x: len(x))
            self.data = data
            
        def proc_URL_TA(self):
            data = self.data.copy()
            # получим все значения g00000 из ссылок
            pattern = re.compile('g\d+')
            data['g_from_URL_TA'] = data['URL_TA'].apply(lambda x: pattern.search(str(x)).group(0))

            # Посчитаем, какое количество городов объединяет в себе каждая группа
            dd = data[['g_from_URL_TA', 'City']].groupby(['g_from_URL_TA', 'City']).size().reset_index()
            dd2 = dd[['g_from_URL_TA', 'City']].groupby('City').count().reset_index()
            dd2.columns = ['City', 'count_g_from_URL_TA']
            data= data.merge(dd2, left_on='City', right_on='City', how='outer')
            data = data.drop(['g_from_URL_TA', 'URL_TA', 'ID_TA'], axis=1)
            self.data = data
            
            
        def get_chain_name(row):
            # переводит имя id в имя группы
            if row['Restaurant_id'] > 1:
                return 'chain_' + str(row['index'])
            else:
                return 'chain_no'
            
        def proc_Id_to_chain(self):
            # переводит id в группу ресторанов
            # считаем сколько ресторанов в группе
            data = self.data.copy()
            d = data['Restaurant_id'].value_counts().reset_index()
            d['Restaurant_chain'] = d.apply(get_chain_name, axis=1)
            data = data.merge(d[['index', 'Restaurant_chain']], left_on='Restaurant_id', right_on='index', how='left')
            data = data.drop(['Restaurant_id', 'index'], axis=1)
            
            d = data['Restaurant_chain'].value_counts().reset_index()
            dd = data[['Restaurant_chain','Rating']][ (data['sample'] == 1)].groupby(by='Restaurant_chain').agg(lambda x:x.value_counts().index[0])

            # посмотрим, различаются ли моды в разных группах, и сколько в них ресторанов
            ddd = dd.merge(d, left_on='Restaurant_chain', right_on='index')
            ddd = ddd[['index', 'Rating','Restaurant_chain']]
            ddd.columns = ['Restaurant_chain', 'mode_of_Rating', 'count_of_Restaurants_in_chain']

            data = data.merge(ddd[['Restaurant_chain','count_of_Restaurants_in_chain']], left_on='Restaurant_chain', right_on='Restaurant_chain',how='left')
            data = data.drop(['Restaurant_chain'], axis=1)
            self.data = data
            
        def get_delta_with_last(row):
            # получает разницу в днях между самым последним отзывом в базе и в ресторане
            days = latest_date
            if (row['count'] > 0):
                days = latest_date - row['dates'][0]
            else:
                if (row['number'] == 0):
                    days = latest_date - earest_date
                else:
                    days = (latest_date - earest_date)/(row['number']+1)
            return days.days

        def get_delta_between_two(row):
            # получает разницу в днях между двумя отзывами
            days = latest_date
            if (row['count'] >= 2):
                days = row['dates'][0] - row['dates'][1]
            elif (row['count'] == 1):
                if (row['number'] == row['count']):
                    days = row['dates'][0] - earest_date
                else:
                    days = (row['dates'][0] - earest_date)/(row['number'])
            else:
                if (row['number'] == 0):
                    days = latest_date - earest_date
                else:
                    days = (latest_date - earest_date)/(row['number']+1)
            return days.days   
        
        
        def proc_dates_of_reviews(self):
            # обработаем Reviews и Number Reviews
            # заполним пропуски
            # найдем разницу в днях между отзывами
            # найдем среднюю частоту публикации по двум отзывам
            
            # новые признаки:
            # data['Reviews_frequence'] = d['frequence'] 
            # data['delta_with_last'] = d['delta_with_last']
            # data['delta_between_two'] = d['delta_between_two']
            
            data = self.data.copy()
            pattern = re.compile('\d\d\/\d\d\/\d\d\d\d')

            s_dates_rews = data['Reviews'].apply(lambda x: (pattern.findall(str(x))))
            s_dates_rews = s_dates_rews.apply(lambda l: [datetime.strptime(d,'%m/%d/%Y') for d in l])
            
            # получим минимальное количество отзывов: 0,1 или 2
            s_min_count_rews = s_dates_rews.apply(lambda l: len(l))

            # найдем даты самого последнего отзыва и самого первого
            latest_date = datetime(2000,1,1,0,0,0)
            for l_d in s_dates_rews:
                if len(l_d)>0:
                    if latest_date<max(l_d):
                        latest_date = max(l_d)

            earest_date = datetime.now()
            for l_d in s_dates_rews:
                if len(l_d)>0:
                    if earest_date>min(l_d):
                        earest_date = min(l_d)            




            d = pd.DataFrame({'number': data['Number of Reviews'],'count': s_min_count_rews, 'dates':s_dates_rews}) 
            # Заполним пропуски Number of Reviews сначала 0, потом заменим 0 числом из Count
            d['number'].fillna(0, inplace=True)
            d['number'] = d.apply(lambda row: row['count'] if row['number']<row['count'] else row['number'], axis=1)
            
            
            d['delta_with_last'] = d.apply(get_delta_with_last, axis=1);
            d['delta_between_two'] = d.apply(get_delta_between_two, axis=1)
            d['frequence'] = d.apply(lambda row: (row['delta_with_last']+row['delta_between_two'])/2, axis = 1)


            data['Reviews_frequence'] = d['frequence']
            data['Number of Reviews'] = d['number']
#             data['delta_with_last'] = d['delta_with_last']
#             data['delta_between_two'] = d['delta_between_two']
            
            data = data.drop(['Reviews'], axis=1)
            
            self.data = data
            

        
        def get_population(row):
            # вручную заполняе пропуски по населению
            x = row['Population']
            city = row['City']
            if city=='Krakow':
                return 779115
            elif city== 'Luxembourg':
                return 124528
            elif city== 'Oporto':
                return 287591
            else:
                return x

        def get_country(row):
            # вручную заполняе пропуски по странам
            x = row['Country']
            city = row['City']
            if city=='Krakow':
                return 'pl'
            elif city== 'Luxembourg':
                return 'lx'
            elif city== 'Oporto':
                return 'pt'
            else:
                return x  
            
        def add_population_and_country(self, df_wc):
            data = self.data.copy()
            df_worldcities = df_wc.copy()
            
            df_worldcities = df_worldcities[~df_worldcities['Population'].isna()][['Country','AccentCity', 'Population']]
            df_worldcities2 = df_worldcities[['AccentCity', 'Population']].groupby('AccentCity').max().reset_index()

            df_worldcities =pd.merge(df_worldcities, df_worldcities2, left_on=['AccentCity', 'Population'], right_on=['AccentCity', 'Population'], how='right')

            # Добавим в основной датасет информацию о странах и населении
            data = data.merge(df_worldcities, left_on='City', right_on='AccentCity', how='left')

            data = data.drop(['AccentCity'], axis=1)

            # Заполним пропуски
            data['Population'] = data[['City', 'Population']].apply(get_population, axis=1)
            data['Country'] = data[['City', 'Country']].apply(get_country, axis=1)

            self.data = data

        def dummy_Cuisines(self):
            data = self.data.copy()
            # применим к кухням One-Hot Encoding
            d = pd.get_dummies(data['Cuisine Style'].apply(pd.Series).stack()).sum(level=0)
            data = pd.concat([data, d], axis=1)
            data = data.drop(['Cuisine Style'], axis=1)
            self.data = data

        def drop_Cuisines(self):
            data = self.data.copy()
            data = data.drop(['Cuisine Style'], axis=1)
            self.data = data            
            
        def dummy_Countries(self):
            data = self.data.copy()
            data = pd.get_dummies(data, columns=[ 'Country',], dummy_na=True)
            self.data = data
            
        def drop_Countries(self):
            data = self.data.copy()
            data = data.drop(['Country'], axis=1)
            self.data = data            
            
        def dummy_Cities(self):
            data = self.data.copy()
            data = pd.get_dummies(data, columns=[ 'City',], dummy_na=True)
            self.data = data
```
